## Assignment 06.01

**Plot some lines**

Write a function ``assignment_06_01`` that draws a line from the point [-1,4] to the point [3,5] with line style ``'r-x'`` and another line from [3,6] to [1,8] with line style ``'k-o'``.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def assignment_06_01():
    plt.plot([1,3],[8,6],'k-o', label="(3,6) to (1,8)");
    plt.plot([-1,3],[4,5], 'r-x', label="(-1,4) to (3,5)")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('fig-01: drow two lines');
    plt.legend()
    
assignment_06_01()    
```

## Assignment 06.02

**Plot a dinosaur**

Write a function ``assignment_06_02`` that draws the data set 'dino' from the tab-separated file ``data/DatasaurusDozen.tsv``. 

```python
def assignment_06_02():
    ds = pd.read_csv('data/DatasaurusDozen.tsv',sep='\t')
    
    x = ds['x'][ds['dataset'] == 'dino']
    y = ds['y'][ds['dataset'] == 'dino']
    
    plt.plot(x, y, 'r.')

    plt.xticks([])
    plt.yticks([])

    plt.title('Fig-02: Dino')
    
assignment_06_02()
```

**Plot some histograms**

Write a function ``assignment_06_03`` creates a normally distributed variable with mean 5 and standard deviation 2. Plot the histogram.

```python
def assignment_06_03():
    normal_var = np.random.normal(5, 2, size=10000)
    plt.hist(normal_var, color='r')
    plt.title('Fig-03: Histogram of normal(5,2)')
assignment_06_03()    
```

## Assignment 06.04

**Plot some 2D Kernel Density Estimates**

Write a function ``assignment_06_04`` creates a 2D normally distributed variable with mean [0,0] and covariance matrix [[5, 2], [2, 2]]. Plot the joint kernel density estimate.


```python
def assignment_06_04():
    mean = [0, 0]
    covariance = [[5, 2], [2, 2]]
    
    x, y = np.random.multivariate_normal(mean, covariance, 1000).T

    g = sns.jointplot(x, y, kind="kde")
    g.fig.tight_layout()
    g.set_axis_labels('X', 'Y')
   
    g.fig.suptitle('Fig-04: 2D normally distributed variable')

assignment_06_04()
```
